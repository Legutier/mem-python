```python
def fetch_fits_data(path):
    hdul = fits.open(path)
    datas = list()
    for i in range(len(hdul)):
        datas.append(hdul[i].data)
    hdul.close()
    return datas
```

```python
def gaussian(x, a, b, c):
    num = -(x - b) ** 2
    den = 2 * c ** 2
    return a * np.exp(num/den)
```

```python
def gaussian2d(x, y, a, b, c):
    z = np.sqrt(x**2 + y**2)
    return gaussian(z, a, b, c)
```

```python
def gaussian_plane(s):
    x = np.linspace(-100, 100, 1000)
    y = np.linspace(-100, 100, 1000)
    xv, yv = np.meshgrid(x, y)
    return gaussian2d(xv, yv, 1, 0, s)
```

```python
data = gaussian_plane(10)
fft = fftshift(fft2(data))
```
